Overall design
"DLPFC (BA9) brain tissues of AD patients, HD patients and non-demented controls samples were obtained from Harvard Brain tissue resource center (HBTRC). The HBTRC samples were primarily of Caucasian ancestry, as only eight non-Caucasian outliers were identified, and therefore excluded for further analysis. Post-mortem interval (PMI) was 17.8+8.3 hours (mean ± standard deviation), sample pH was 6.4±0.3 and RNA integrity number (RIN) was 6.8±0.8 for the average sample in the overall cohort. Tissues were profiled on a custom-made Agilent 44K array (GPL4372). 624 individual DLPFC samples were profiled against a common DLPFC pool constructed from the same set of samples."

What makes this dataset an interesting example:
    1- Alzheimer Disease is mostly sporadic (not genetic), while Hungtington's Disease is purely 
        genetic. Hungtington's Disease is caused by a dominant mutation, so the predictability of 
        Hungtington's Disease is already 100% accurate. The "development" of other neurodegenrative
        disorders are also thought to be influenced by genetic (but not that much; a
        small proportion are "family cases", which are also caused by dominant mutations).  
        
Hungtington disease, in neurogenerative disorders, is thus the only one with a truly certain diagnostic. Alzheimer's disease may be confused with other resembling diseases (other dementias). The probable variaty of causes for the pathogenesis (genetic vs sporadic for example), one might speculate Alzheimer's disease would be a way of describing (or classifying) people with similar phenotypes (symptomes) rather than describing a common disease. The incidence of Alzheimer's disease increase a lot with higher age. It might be possible that given enough time, everyone could go through an Alzheimer-like disease, but die before that time. (might make sense to try predicting the age of onset, including never-seen-before, ages such as 150 years old, with the help of Bayesian inferences). 

This is working fine with the datset presented, but it might be interesting to try with negative samples (for both the diseases) that were not selected to make good controls. Especially with old age that has other types of dementias. I wonder if AD would be so easy to distinguish from. Here, HD is interesting because it can be distinguished from AD, which makes a good example, but again, HD is expected to be the most distinguishable neurodegenrative disorder, but others might not be. It could help to support my hypothesis that similar phenotypes can be explained by more than one cause, and vice-versa (one cause can induce more than one phenotype, because phenotypes are plastic, some more than others).

https://stats.stackexchange.com/questions/321841/what-are-variational-autoencoders-and-to-what-learning-tasks-are-they-used

Ideal parameters:
$$ \min_{\boldsymbol{\lambda}}\mathcal{D}[p(\mathbf{z}\vert \mathbf{x})\vert\vert q(\mathbf{z}\vert \mathbf{x},\boldsymbol{\lambda})] $$

It should also minimize the reconstruction loss (and optional regularization terms, mainly L1 or L2)

VAE loss function:
$$ELBO(\boldsymbol{\lambda})= E_{q(\boldsymbol{z}\vert \mathbf{x},\boldsymbol{\lambda})}[\log p(\mathbf{x}\vert\boldsymbol{z})]-\mathcal{D}[(q(\boldsymbol{z}\vert \mathbf{x},\boldsymbol{\lambda})\vert\vert p(\boldsymbol{z})]$$

$$ q(\mathbf{z}\vert \mathbf{x},\boldsymbol{\lambda}) = \mathcal{N}(\mathbf{z}\vert\boldsymbol{\mu}(\mathbf{x}), \boldsymbol{\sigma}^2(\mathbf{x})I) $$

conditional distribution:
$$ p_{\boldsymbol{\phi}}(\mathbf{x}\vert\mathbf{z}) = \mathcal{N}(\mathbf{x}|\boldsymbol{\mu}(\mathbf{z}; \boldsymbol{\phi}), \boldsymbol{\sigma}(\mathbf{z}; \boldsymbol{\phi})^2I)$$


$$ ELBO(\boldsymbol{\theta},\boldsymbol{\phi})= \sum_i E_{q_{\boldsymbol{\theta}}(\boldsymbol{z}\vert \mathbf{x}_i,\boldsymbol{\lambda})}[\log p_{\boldsymbol{\phi}}(\mathbf{x}_i\vert\boldsymbol{z})]-\mathcal{D}[(q_{\boldsymbol{\theta}}(\boldsymbol{z}\vert \mathbf{x}_i,\boldsymbol{\lambda})\vert\vert p(\boldsymbol{z})] $$
